```python
import csv
import numpy as np
import matplotlib.pyplot as plt

base_filename = 'resnet16_eps0.1_benchmark'

counts = [5, 10, 50, 100, 200]

for count in counts:
    benchmark_file = f'{base_filename}_{count}.csv'
    with open(benchmark_file, 'r') as f:
        reader = csv.reader(f)
        data = list(reader)

    data = np.array(data).astype(float)
    data[:, 1] = 2 ** data[:, 1]
    values, counts = np.unique(data[:, 1], return_counts=True)
    plt.bar(values.astype(np.int16).astype(np.str_), counts)
    plt.title(f'MCTS Iterations: {count}')
    plt.show()
```
